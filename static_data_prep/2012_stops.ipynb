```python
import pandas as pd
from pyproj import Proj, transform
import utm
import convertbng
from convertbng.util import convert_lonlat
```

```python
df = pd.read_csv('2012_stops.csv', encoding='latin-1', header=0, index_col=False)
```

```python
df.head(50)
```

```python
df = df[pd.notnull(df['NaPTANId'])]
```

```python
df_out = df[df['NaPTANId'].str.contains("DB")]
df_out.head(10)
```

```python
df_out.shape
```

```python
def proj_transform(df):
    prj_itm = Proj(init='epsg:2157')
    wgs84 = Proj(init='epsg:4326')
    lon,lat = transform(prj_itm,wgs84,df['Easting'].tolist(),df['Northing'].tolist())
    
    df['lat'] = lat
    df['lon'] = lon
    return df
```

```python
df_out = proj_transform(df_out)
```

```python
df_out.head(100)
```

```python
df_out['NaPTANId'] = df_out['NaPTANId'].apply(lambda x: x.split('DB00')[-1])
```

```python
df_out.head(100)
```

```python
df_out['NaPTANId'] = df_out['NaPTANId'].str.lstrip('0')
```

```python
df_out.head(100)
```

```python
df_out.rename(columns = {'NaPTANId':'Stop No'}, inplace = True)
```
