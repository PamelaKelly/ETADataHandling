# Drop Some features we don't need 

```python
#Drop duplicate rows 
df=df.drop_duplicates()
#Drop constant column
df.drop(['Direction','Operator'],axis=1,inplace=True)
```

```python
#Drop the features that we don't need 
df.drop(['Next_Timestamp','Between_Time'],axis=1,inplace=True)
```

```python
#Drop Unused column
df.drop(['Line_ID','Congestion','Block_ID','Vehicle_ID'],axis=1,inplace=True)
```

```python
#Check how many rows we deleted(Before it is  (9204432, 18))
df.shape
# The result is (9204432,10)
```

```python
#Show the first 5 rows.
df
```

# Do more cleaning for the data frame 

```python
df_groupby=df.groupby(['Vehicle_Journey_ID','Date','Journey_Pattern_ID'])

#size = df_groupby.size()
#size=size[size<13]

#keys=df_groupby.groups
```

```python
# Filter all the data only have less than 13 rows data 
df_short_Journ=df_groupby.filter(lambda x: len(x) < 13)
```

```python
# Drop all the Journery that are less than a number 

df=pd.concat([df, df_short_Journ]).drop_duplicates(keep=False)
```

```python
#Check how many rows we deleted(Before it is(9204432, 10)
df.shape
# The result now is (9166233, 10)
```

```python
%time df.groupby(['Vehicle_Journey_ID','Journey_Pattern_ID','Date']).size().reset_index().rename(columns={0:'count'})
```

```python
# Only keep the first row for every trip at the same Stop(keep both at or not_at stop)
df = df.drop_duplicates(['Vehicle_Journey_ID','Journey_Pattern_ID','Date','Stop_ID','At_Stop'])
```

```python
# Check how many rows are deleted before it is (9134097, 13)
df.shape 
# The result now is (3175832, 10)
```

# Add new features for the data frame 

```python
#Convert Timestamp to datetime and new column datetime to data frame 
df['datetime'] = pd.to_datetime(df['Timestamp'],unit='s')
df.head(5)
```

```python
#ADD day of the week to the date frame 
df['day_of_week'] = df['datetime'].dt.weekday_name
df.head(3)
```

```python
# Add new feature Trip_Time 
zscore=lambda x: (x - x.min())
df['Trip_Time']= df.groupby(['Vehicle_Journey_ID','Journey_Pattern_ID','Date'])['Timestamp'].transform(zscore)
```

```python
#Change date column type for the convertion to day of week 
#df['Date'] = pd.to_datetime(df['Date'])
#pd.to_datetime(df['Date'], format="%m/%d/%Y")
df.dtypes
```

# Descibe  categorical_columns and continuous_columns features 

```python
# Select columns containing categorical data
categorical_columns = df[['Journey_Pattern_ID','Vehicle_Journey_ID','Stop_ID','At_Stop','day_of_week']].columns

# Convert data type to category for these columns
for column in categorical_columns:
    df[column] = df[column].astype('category')
# Descriptive stats for categorical features
#df.select_dtypes(['category']).describe().T
df[categorical_columns].describe().T
```

```python
#Find all the continuous features 

continuous_columns=df[['Lon_WGS84','Lat_WGS84','Delay','Distance_Terminal','Trip_Time']].columns

# Descriptive stats for continuous features
df[continuous_columns].describe().T
```
