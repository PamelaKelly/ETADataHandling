# Train model for  every Journey_Pattern_ID

```python
df.shape
```

```python
# Train model for every Journey Pattern ID 

def train_model(df):
    df_grouped = df.groupby(['Journey_Pattern_ID'])
    max_depth = 30

    for journey_name, journey_group in df_grouped:
        regr_rf = RandomForestRegressor(max_depth=max_depth,random_state=2)
    
    # Prepare the descriptive features
        X = pd.concat([journey_group[['Distance','midweek','HourOfDay','rain']]], axis=1)
        y = journey_group.Trip_Time 
        regr_rf.fit(X,y)
        score_rf=regr_rf.score(X,y)
        mae = abs(y-regr_rf.predict(X)).mean()
    
        print(journey_name,' score is ',score_rf, 'Mean Absolute error is',mae)
        pickle.dump(regr_rf, open('models_every_jour/{}model.sav'.format(journey_name), 'wb'))
        
    return None
%time train_model(df)
```
